```python
import numpy as np
import pandas as pd 
from pandas.api.types import is_numeric_dtype
```

```python
data = pd.read_csv("./burger_king/burger-king-menu.csv")
```

```python
data.head()
```

```python
m = data.isnull().sum()
```

```python
m
```

```python
for n,c in data.items():
    if is_numeric_dtype(c):
        if data[n].isnull().sum():
            print(n,data[n].isnull().sum())
```

```python
for n, c in data.items():
    if is_numeric_dtype(c):
        if data[n].isnull().sum():
            data[n+"_na"]=pd.isnull(c)
        data[n] = c.fillna(c.median())
    print(n,data[n].isnull().sum())
        
        
```

```python
data
```

```python
print(data["Category"].value_counts())
```

```python
replacing = {"Category" : {"Breakfast" : 1,
                           "Burgers" : 2,
                           "Chicken" : 3}}
```

```python
data_replace = data.copy()
data_replace.replace(replacing, inplace = True)
data_replace
```

```python
X = data_replace.iloc[:,0:8]
y = data_replace.iloc[:,8]
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.25, random_state=0)
```

```python
final = {"X train " : X_train,
         "X validation " : X_val, 
         "X test " : X_test,
         "y train " : y_train,
         "y validation " : y_val,
         "y test " : y_test}
for i in final:
    print(f"number of {i} : {len(final.get(i))}")
```

### note : have to solve y train value

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt


rf = RandomForestRegressor(n_estimators=5)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

trs = rf.score(X_train, y_train)
ts = rf.score(X_test, y_test)
rmse = sqrt(mean_squared_error(y_test, y_pred))
    

```

```python
print("training score :", trs)
print("test score :" , ts)
print("root mean squared error :", rmse)
```
